## Find the contacts between ions and proteins

### Define the function

```python
def filter_residues(traj, query_indices: "list[uint64]", cutoff: "float" = 2.0):
    """ Determine residues whose at least one atom comes within cutoff distance of the query, frame by frame

    Arguments
    ----------------
    traj -- MDTraj Trajectory object
    query_indices (list[int]) -- indices of atoms around which the selection is to be expanded 
    NB the atom indices must follow the MDTraj Topology object's atoms' 'index' property, 
    NOT the 'serial' property
    cutoff (float) -- cutoff distance (nm) for computing neighbouring atoms (default value is 2.0 nm)

    Returns
    ----------------
    neighbour_residues (list[list[int]]) -- list of lists of residue sequence numbers (following teh MDTraj Topology object's
    residues' 'resSeq' property, NOT the 'index' property). For a trajectory of n frames, the neighbour_residues 'master list'
     will have length n and its ith element will correspond to the ith frame of the trajectory
    """

    # Compute neighbours within a set cutoff distance from the selected atom set
    neighbours = md.compute_neighbors(
        traj=traj, cutoff=cutoff, query_indices=query_indices
    )

    # Get the topology object
    top = traj.topology

    # Create an empty list to be populated with lists of residues
    neighbour_residues = []

    # Iterate over frames
    for frame in neighbours:
        # Create an empty list to be populated with residue sequence numbers for this frame
        neighbour_residues_f = []
        # Iterate over residues in the topology
        for residue in top.residues:
            # Iterate over atoms in each residue
            for atom in residue.atoms:
                # If an atom was found to be in a residue within cutoff distance from query in this frame,
                # add residue sequence number to neighbour_residues_f and move on to the next residue
                if atom.index in frame and atom.index not in query_indices:
                    if residue.resSeq not in neighbour_residues_f:
                        neighbour_residues_f.append(residue.resSeq)
                        break
        # For each frame, once the list of residues is complete, add this list to the 'master list'
        neighbour_residues.append(neighbour_residues_f)

    return neighbour_residues
```

## Find Na+ ion contacts

```python
# ~10 minutes to run
Na_Neighbor = filter_residues(HPI_traj, query_indices = Na_ions, cutoff=0.25)
```

```python
# Collapse list of lists into single list
Merged_Na_Neigh = list(itertools.chain(*Na_Neighbor))

# Count values and convert to dictionary
Counter_Na_Neigh = Counter(Merged_Na_Neigh)
```

```python
# Highest Residue Number with contact to an Ion
max(Merged_Na_Neigh)
```

```python
# Residue with most contacts
dict(Counter(Counter_Na_Neigh).most_common(5))
```

```python
# Plot
keys = Counter_Na_Neigh.keys()
values = Counter_Na_Neigh.values()

plt.figure(figsize=(20,7.5))
plt.bar(keys, values)
#plt.xlim(-101, 16000)
```

### Preparing dictionary

```python
# Read in Excel data containing the "residue monomer number" to each residue number in the 60-mer
df_resi_no = pd.read_excel('Residue_Number_excel.xlsx', sheet_name='Sheet4', header=None, index_col=0)
df_resi_no.head()
```

```python
# Transpose and return a dictionary of lists with Key=Column name and Value=List
dict_resi_no = df_resi_no.T.to_dict('list')
```

### Na Dictionary

```python
# Dictionary "residue monomer number" to contact numbers
final_dict_Na = {}
for res_n in dict_resi_no.keys():
	curr_contacts = dict_resi_no[res_n]
	curr_sum = 0
	for c in curr_contacts:
		curr_sum += Counter_Na_Neigh[c]
	final_dict_Na[res_n] = curr_sum
```

```python
# Most common residue contacts
dict(Counter(final_dict_Na).most_common(5))
```

```python
# Get residue Mean
dict_Na_divided_30 = {k: v / 30 for k, v in final_dict_Na.items()}
# Most common residue contacts
dict(Counter(dict_Na_divided_30).most_common(5))
```

### Plot

```python
# Choose what of the 265 positions to plot index = 0:264
x = dict(Counter(dict_Na_divided_30).most_common(25)).keys()
y = dict(Counter(dict_Na_divided_30).most_common(25)).values()

# Plot
fig = plt.figure()
ax = fig.add_subplot(111)
#ax.set_xlim(-2, 266)
ax.set_ylim(0, 45)

ax.set_xlabel('Residue', fontsize=12)
# Number of residues having a Na+ ion closer than 2.5 Å over 20 ns, (n=30)
ax.set_ylabel('Number of Sodium Ion Contacts', fontsize=12)
plt.xticks(rotation=90, fontsize = 9)

bar = ax.bar(x,y, color='mediumblue', label='Na\u207A')

ax.legend(handles=[bar], fontsize=10)

plt.savefig('Contacts_Na-Leu.png', bbox_inches="tight")
```
