# Imports

```python
import mdtraj as md
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data in

```python
# Load the protein only trajectory
traj = md.load_dcd(("ion_output_corrected_-1.dcd"), top='Res_Last_Frame.pdb')
```

```python
# Could not get protein COM to function in the next step, negligible ions compared to protein atoms
system_COM = md.compute_center_of_mass(traj)
plt.plot(system_COM)
```

## Distances from COM

### COM to protein atoms

```python
# selecting protein as index
protein_selection = traj.topology.select('protein')
print(protein_selection)
```

```python
#calculate masses of all atoms:
masses = np.array([i.element.mass for i in traj.topology.atoms])
```

```python
#iterate through all frames and save the distances of each atom to the centre of mass:
d = list()
for frame in traj.xyz:
    protein_coords = frame[protein_selection]
    #COM of whole system:
    centre_of_mass = np.average(frame, axis=0, weights=masses)
    #Compute vectos to distances and compile:
    vectors = (protein_coords - centre_of_mass)
    distances = np.linalg.norm(vectors, axis=1)
    d += list(distances)
```

```python
print(type(d),
      len(d),
      "frames = 100",
      "total protein atoms = 258780",
      "len(d)/100 =",
      len(d)/100,
      np.average(d),
      np.max(d),
      np.min(d),
     )
```

```python
rborder = np.average(d)
print(rborder)
```

```python
##Take a histogram:
Pcounts, Plengths = np.histogram(d,bins=250)
#normalize 'counts' by averaging across the number of trajectory frames:
Pcounts = Pcounts / len(traj)
#calculate the volume of each spherical shell:
Pshell_volumes = 4/3*np.pi* ( Plengths[1:]**3 - Plengths[:-1]**3 )
#normalize 'counts' by the volume of each shell:
Pcounts = Pcounts / Pshell_volumes
```

```python
#plot:
plt.plot( (Plengths[:-1]+Plengths[1:])/2, 
           Pcounts , '-o' )
plt.ylabel('Number density per nm^3')
plt.xlabel('r (nm)')
plt.axvline(rborder, c='r', label='Protein average radius')
plt.legend()
```

### COM to Sodium Ions

```python
#select sodium ions indices:
Na = traj.top.select('symbol == Na')
# Number of sodium ions:
No_Na = len(Na)
```

```python
#iterate through all frames and save the distances of each atom to the centre of mass:
n = list()
for frame in traj.xyz:
    Na_coords = frame[Na]
    #COM of whole system:
    centre_of_mass = np.average(frame, axis=0, weights=masses)
    #Compute vectos to distances and compile:
    vectors = (Na_coords - centre_of_mass)
    distances = np.linalg.norm(vectors, axis=1)
    n += list(distances)
```

```python
print(type(n),
      len(n),
      "frames = 100",
      "total sodium atoms =",
      No_Na,
      "len(n)/100 =",
      len(n)/100,
      np.average(n),
      np.max(n),
      np.min(n),
     )
```

```python
##Take a histogram:
Na_counts, Na_lengths = np.histogram(n,bins=150)
#normalize 'counts' by averaging across the number of trajectory frames:
Na_counts = Na_counts / len(traj)
#calculate the volume of each spherical shell:
Na_shell_volumes = 4/3*np.pi* ( Na_lengths[1:]**3 - Na_lengths[:-1]**3 )
#normalize 'counts' by the volume of each shell:
Na_counts = Na_counts / Na_shell_volumes
```

```python
#plot:
plt.plot( (Na_lengths[:-1]+Na_lengths[1:])/2, 
           Na_counts , '-o' )
plt.ylabel('Number density per nm^3')
plt.xlabel('r (nm)')
plt.axvline(rborder, c='r', label='Protein average radius')
plt.legend()
```

### COM to Chloride Ions

```python
#select sodium ions indices:
Cl = traj.top.select('symbol == Cl')
# Number of chloride ions
No_Cl = len(Cl)
```

```python
#iterate through all frames and save the distances of each atom to the centre of mass:
c = list()
for frame in traj.xyz:
    Cl_coords = frame[Cl]
    #COM of whole system:
    centre_of_mass = np.average(frame, axis=0, weights=masses)
    #Compute vectos to distances and compile:
    vectors = (Cl_coords - centre_of_mass)
    distances = np.linalg.norm(vectors, axis=1)
    c += list(distances)
```

```python
print(type(c),
      len(c),
      "frames = 100",
      "total chloride atoms =",
      No_Cl,
      "len(c)/100 =",
      len(c)/100,
      np.average(c),
      np.max(c),
      np.min(c),
     )
```

```python
##Take a histogram:
Clcounts, Cllengths = np.histogram(c,bins=150)
#normalize 'counts' by averaging across the number of trajectory frames:
Clcounts = Clcounts / len(traj)
#calculate the volume of each spherical shell:
Clshell_volumes = 4/3*np.pi* ( Cllengths[1:]**3 - Cllengths[:-1]**3 )
#normalize 'counts' by the volume of each shell:
Clcounts = Clcounts / Clshell_volumes
```

```python
#plot:
plt.plot( (Cllengths[:-1]+Cllengths[1:])/2, 
           Clcounts , '-o' )
plt.ylabel('Number density per nm^3')
plt.xlabel('r (nm)')
plt.axvline(rborder, c='r') # label='Protein average radius'
plt.legend()
```

## Flow OR Ions transited (crossed) total

#### Change list to array and re-shape frame to each SODIUM ion

```python
Na_Dist_1 = np.asarray(n).reshape((No_Na, 100))
df_Na_Dist_1 = pd.DataFrame(Na_Dist_1)
Na_count_1 = df_Na_Dist_1[df_Na_Dist_1 < rborder].count()
plt.plot(Na_count_1)
```

#### Change list to array and re-shape frame to each CHLORIDE ion

```python
Cl_Dist_1 = np.asarray(c).reshape((No_Cl, 100))
df_Cl_Dist_1 = pd.DataFrame(Cl_Dist_1)
Cl_count_1 = df_Cl_Dist_1[df_Cl_Dist_1 < rborder].count()
plt.plot(Cl_count_1)
```

#### Plot

```python
timens = np.arange(0, 20, 0.2)

fig = plt.figure()
ax = fig.add_subplot(111)
#ax.set_xlim(-2.5, 27.5)
#ax.set_ylim(-25, 600)

ax.set_xlabel('Time (ns)', fontsize=12)
ax.set_ylabel('Number of Ions Crossed', fontsize=12)

line_up, = plt.plot(timens, Na_count_1.diff().abs().cumsum(), c='mediumblue', label='Na\u207A')
line_down, = plt.plot(timens, Cl_count_1.diff().abs().cumsum(), c='lime', label='Cl\u207B')

ax.legend(handles=[line_up, line_down], fontsize=10)

plt.savefig('Na-Cl-Flow-Leu.png', bbox_inches="tight")
```

## Gradients

```python
# Line from graph above
Na_line = Na_count_1.diff().abs().cumsum().fillna(0)
Cl_line = Cl_count_1.diff().abs().cumsum().fillna(0)
```

```python
# Total Ion Flow
print("Na", Na_line.max(), "Cl", Cl_line.max())
```

```python
# Gradient for each 5 ns of simulation divide by 5 for 1 ns flow, averaged and standard dev

slope_Na_5, intercept_Na_5 = (np.polyfit(timens[0:25], Na_line[0:25], 1))/5

slope_Na_10, intercept_Na_10 = (np.polyfit(timens[25:50], Na_line[25:50], 1))/5

slope_Na_15, intercept_Na_15 = (np.polyfit(timens[50:75], Na_line[50:75], 1))/5

slope_Na_20, intercept_Na_20 = (np.polyfit(timens[75:100], Na_line[75:100], 1))/5

Na_slopes = (slope_Na_5, slope_Na_10, slope_Na_15, slope_Na_20)

print(np.mean(Na_slopes), '±', np.std(Na_slopes), 'mean ± std (n = 4)')
```

```python
# Gradient for each 5 ns of simulation divide by 5 for 1 ns flow, averaged and standard dev

slope_Cl_5, intercept_Cl_5 = (np.polyfit(timens[0:25], Cl_line[0:25], 1))/5

slope_Cl_10, intercept_Cl_10 = (np.polyfit(timens[25:50], Cl_line[25:50], 1))/5

slope_Cl_15, intercept_Cl_15 = (np.polyfit(timens[50:75], Cl_line[50:75], 1))/5

slope_Cl_20, intercept_Cl_20 = (np.polyfit(timens[75:100], Cl_line[75:100], 1))/5

Cl_slopes = (slope_Cl_5, slope_Cl_10, slope_Cl_15, slope_Cl_20)

print(np.mean(Cl_slopes), '±', np.std(Cl_slopes), 'mean ± std (n = 4)')
```
